```python
##preprocessing in single sample
sample=reviews['review'][0]
sample
```

step1:removing html tags


```python
from bs4 import BeautifulSoup##parse html
```

```python
def remove_tags(text):
  html= BeautifulSoup(text)
  return html.text
```

```python
sample_no_tags=remove_tags(sample)
sample_no_tags
```

step2:removing punctuations

```python
import re ##regular exp
```

```python
def remove_punctuations(text):
  return re.sub('[^a-zA-Z]', ' ', text)
```

```python
sample_no_punc=remove_punctuations(sample_no_tags)
sample_no_punc
```

step 3:removing whitespaces

```python
def remove_spaces(text):
  return re.sub(r'\s+',' ',text)
```

```python
sample_no_spaces = remove_spaces(sample_no_punc)
sample_no_spaces
```

step 4:remove single characters

```python
def remove_singles(text):
  return re.sub(r'\b[a-zA-Z]\b','',text)
```

```python
sample_clean=remove_singles(sample_no_spaces)
sample_clean
```

```python
sample_clean=remove_spaces(sample_clean)
sample_clean
```

```python
##preprocessed data into a func
def clean_text(text):
  text=remove_tags(text)
  text=remove_punctuations(text)
  text=remove_singles(text)
  text=remove_spaces(text)
  return text

```

```python
sample
```

```python
sample_clean2=clean_text(sample)
sample_clean2
```

```python
##applying to all reviews
clean_reviews=reviews['review'].apply(func=clean_text)
```
