step1:removing html tags


```python
from bs4 import BeautifulSoup##parse html
```

```python
def remove_tags(text):
  html= BeautifulSoup(text)
  return html.text
```

```python
sample_no_tags=remove_tags(sample)
sample_no_tags
```

step2:removing punctuations

```python
import re ##regular exp
```

```python
def remove_punctuations(text):
  return re.sub('[^a-zA-Z]', ' ', text)
```

```python
sample_no_punc=remove_punctuations(sample_no_tags)
sample_no_punc
```

step 3:removing whitespaces

```python
def remove_spaces(text):
  return re.sub(r'\s+',' ',text)
```

```python
sample_no_spaces = remove_spaces(sample_no_punc)
sample_no_spaces
```

step 4:remove single characters

```python
def remove_singles(text):
  return re.sub(r'\b[a-zA-Z]\b','',text)
```

```python
sample_clean=remove_singles(sample_no_spaces)
sample_clean
```

```python
sample_clean=remove_spaces(sample_clean)
sample_clean
```

```python
##preprocessed data into a func
def clean_text(text):
  text=remove_tags(text)
  text=remove_punctuations(text)
  text=remove_singles(text)
  text=remove_spaces(text)
  return text

```

```python
sample
```

```python
sample_clean2=clean_text(sample)
sample_clean2
```

```python
##applying to all reviews
clean_reviews=reviews['review'].apply(func=clean_text)
```

```python
##explore vector file
glove=pd.read_csv('glove.6B.100d.txt',sep=' ',quoting=3, header=None, index_col=0)
```

```python
print(glove.index.values.take(list(range(0,100,10))))
print(glove.index.values[:10])
```

```python
##creating embedding matrix
```

```python
embeddings_dictionary={key:val.values for key,val in glove.T.items()}
```

```python
embeddings_dictionary
```

```python
embeddings_matrix=np.zeros((SIZE,100))

for word,index in tokenize.word_index.items():
  emb_vector=embeddings_dictionary.get(word)
  if emb_vector is not None:
    embeddings_matrix[index]=emb_vector
```

DEEPLEARNING

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import Dense,Flatten,Activation
from tensorflow.keras.layers import Dropout,GlobalMaxPool2D
```

Simple fully connected NN

```python
model=Sequential()
```

```python
###first layer
emb_layer=Embedding(
    input_dim=SIZE,  ##tot num of input sequence
    output_dim=100,  ##size of output
    weights=[embeddings_matrix],  ##gets weight vector from emb matrix
    input_length=pad_length,  ##len of single sequence=150
    trainable=False  ##make it untrainable so weights dont change while training
)
```

```python
model.add(emb_layer)
model.add(Flatten())
```

```python
##adding o/p dense layer
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
##compile
model.compile(
    optimizer='sgd',
    loss='binary_crossentropy',  ##binary classification
    metrics=['accuracy']
)
```

```python
##train
history=model.fit(
    x_train,
    y_train,
    batch_size=128,
    epochs=20,
    validation_split=0.25
)
```

```python
score=model.evaluate(x_test,y_test,return_dict=True)
score
```

```python
## function to plot loss and accuracy
def plot_history(history):
  fig = plt.figure(figsize=(10, 4))
## accuracy plot
  plt.subplot(1, 2, 1)
  plt.plot(history.history['accuracy'])
  plt.plot (history.history['val_accuracy'])
  plt.title('Accuracy of Model')
  plt.xlabel( 'Epochs')
  plt.ylabel( 'Accuracy' )
  plt.legend(['train', 'test'], loc='lower right')

## loss plot
  plt.subplot(1, 2, 2)
  plt.plot(history.history[ 'loss' ] )
  plt.plot(history.history[ 'val_loss' ] )
  plt.title( 'Loss of Model')
  plt.xlabel( 'Epochs')
  plt.ylabel( 'Loss')
  plt.legend(['train', 'test '], loc= 'upper right')

  plt.tight_layout()
  fig.show();
```

```python
plot_history(history)
```

USING CONVULATIONAL NN

```python
from tensorflow.keras.layers import Conv1D,GlobalMaxPool1D
```

```python
model=Sequential()
```

```python
##embedding layer
emb_layer=Embedding(
    input_dim=SIZE,
    output_dim=100,  ##size of output
    weights=[embeddings_matrix],  ##gets weight vector from emb matrix
    input_length=pad_length,  ##len of single sequence=150
    trainable=False  ##make it untrainable so weights dont change while training
)
```

```python
model.add(emb_layer)
```

```python
model.add(Conv1D(
    filters=64,
    kernel_size=6,
    activation='relu'
))
```

```python
model.add(GlobalMaxPool1D())
```

```python
##output layer
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(
    optimizer='sgd',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
###train
history=model.fit(
    x_train,
    y_train,
    batch_size=128,
    epochs=20,
    validation_split=0.25
)
```

```python
plot_history(history)
```

```python
score=model.evaluate(x_test,y_test,return_dict=True)
score
```
